```python
import pandas as pd
df = pd.read_csv(r'Titanic.csv')
```

**Упражнение 1.** Найдите средний возраст пассажиров

```python
mean_age = df['Age'].mean()
print(f"Средний возраст пассажиров: {mean_age:.2f} лет")
```

**Упражнение 2.** Выведите имя самого молодого пассажира.

```python
min_age = df['Age'].min()
youngest = df[df['Age'] == min_age]
print(f"Самый молодой пассажир (возраст {min_age} лет):")
print(youngest['Name'].to_string(index=False))
```

**Упражнение 3.** Посчитайте доли пассажиров по классам обслуживания (Pclass)

```python
pclass_shares = df['Pclass'].value_counts(normalize=True).sort_index()
print("Доли пассажиров по классам:")
print(pclass_shares)
```

**Упражнение 4.** Найдите средние возраста пассажиров по классам обслуживания.

```python
mean_age_by_class = df.groupby('Pclass')['Age'].mean()
print("Средний возраст по классам:")
print(mean_age_by_class)
```

**Упражнение 5.** Удалите столбец с наибольшим числом пропусков.

```python
null_counts = df.isnull().sum()
max_null_col = null_counts.idxmax()
max_null_count = null_counts.max()
print(f"Столбец '{max_null_col}' имеет {max_null_count} пропусков. Удаляем.")
df.drop(columns=[max_null_col], inplace=True)
```

**Упражнение 6.** Удалите столбцы с наибольшим числом уникальных значений.

```python
unique_counts = df.nunique()
max_unique = unique_counts.max()
cols_to_drop = unique_counts[unique_counts == max_unique].index.tolist()
print(f"Столбцы с наибольшим числом уникальных значений ({max_unique}): {cols_to_drop}")
df.drop(columns=cols_to_drop, inplace=True)
```

**Упражнение 7.** Найдите доли выживших среди мужчин, среди женщин и среди всех пассажиров.

```python
survival_men = df[df['Gender'] == 'male']['Survived'].mean()
survival_women = df[df['Gender'] == 'female']['Survived'].mean()
survival_all = df['Survived'].mean()
print("Доли выживших:")
print(f"  Среди мужчин:   {survival_men:.2%}")
print(f"  Среди женщин:   {survival_women:.2%}")
print(f"  Среди всех:     {survival_all:.2%}")
```

**Упражнение 8.** Найдите доли выживших по классам обслуживания.

```python
survival_by_class = df.groupby('Pclass')['Survived'].mean()
print("Доли выживших по классам:")
print(survival_by_class)
```

**Упражнение 9.** Найдите доли выживших среди возрастных категорий 18-, 18-60, 60+.

```python
bins = [0, 18, 60, 120]
labels = ['<18', '18-60', '60+']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
survival_by_age = df.groupby('AgeGroup')['Survived'].mean()
print("Доли выживших по возрастным группам:")
print(survival_by_age)
```

**Упражнение 10.** Любым из рассмотренных методов реализуйте left join, right join, inner join, outer join датафреймов df1 и df2 (объединение по полю A).

```python
df1 = pd.DataFrame({
    'A': ['1', '2', '4', '7', '10'],
    'B': ['ab', 'c', 'bc', 'ab', 'efd'],
    'C': [4, 4, 3.5, 15, 5]})

df2 = pd.DataFrame({
    'A': ['1', '2', '4', '7', '10'],
    'F': ['n22', 'c1', 'b8', 'a589', 'b95'],
    'D': [100, 400, 30, 150, 50]})
print("Left join:\n", pd.merge(df1, df2, on='A', how='left'), "\n")
print("Right join:\n", pd.merge(df1, df2, on='A', how='right'), "\n")
print("Inner join:\n", pd.merge(df1, df2, on='A', how='inner'), "\n")
print("Outer join:\n", pd.merge(df1, df2, on='A', how='outer'))
```
